```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
# Load cortical positions - Comment/Remove to generate new positions
Cortical_Neuron_xy_Positions = np.loadtxt("cortical_xy_pos.txt", delimiter=",")
print(Cortical_Neuron_xy_Positions.shape)

plt.plot(Cortical_Neuron_xy_Positions[0], Cortical_Neuron_xy_Positions[1], '*')
plt.plot([0, 0], [-1750, 1750], '*', label='recording electrode positions on xy plane')
plt.plot([0], [0], '*', label='stimulating electrode position (on a plane 4.075cm below)')
plt.title('Cortical Neuron population on the xy plane')
plt.xlabel('X (m)')
plt.ylabel('Y (m)')
plt.xlim([-3500, 3500])
plt.ylim([-3500, 3500])
plt.legend()
plt.show()   
```
