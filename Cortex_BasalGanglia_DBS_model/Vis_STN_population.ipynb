```python
import numpy as np
import matplotlib.pyplot as plt

STN_Neuron_xy_Positions = np.loadtxt("STN_xy_pos_new.txt", delimiter=",")
print(STN_Neuron_xy_Positions.shape)

plt.plot(STN_Neuron_xy_Positions[0], STN_Neuron_xy_Positions[1], '*')
plt.plot([0, 0], [-1500, 1500], '*', label='orginal recording positions on xy plane')
plt.plot([0], [0], '*', label='stimulating electrode position (on a plane 250µm below)')
plt.title('STN Neuron population on the xy plane')
plt.xlabel('X (µm)')
plt.ylabel('Y (µm)')
plt.xlim([-3500, 3500])
plt.ylim([-3500, 3500])
plt.legend()
plt.show()  
```
