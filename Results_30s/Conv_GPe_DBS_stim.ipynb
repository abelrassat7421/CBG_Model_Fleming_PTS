```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
interp_DBS_amplitudes = np.array(
    [0.0, 0.25, 0.5, 0.75, 1.0, 1.25, 1.5, 1.75, 2, 2.25, 2.50, 3, 4, 5]
)
interp_collaterals_entrained = np.array(
    [0, 0, 0, 1, 4, 8, 19, 30, 43, 59, 82, 100, 100, 100]
)

plt.plot(interp_DBS_amplitudes, interp_collaterals_entrained, "*")
plt.title("Number of entrained collaterals versus stimulation amplitude")

# Need to write a dictionnary with all stimulation values from 0 to 4.0 with step 0.1 linked to the corresponding number of entrained collaterals
```

```python
DBS_amplitudes = np.arange(0, 4.1, 0.1)
collaterals_entrained = [np.interp(DBS_amp, interp_DBS_amplitudes, interp_collaterals_entrained) for DBS_amp in DBS_amplitudes]

plt.plot(DBS_amplitudes, collaterals_entrained, '*')
plt.title("Number of entrained collaterals versus stimulation amplitude smooth")


print("Number of entrained collaterals for DBS amplitude of 0.8 mA: ", np.interp(0.8, interp_DBS_amplitudes, interp_collaterals_entrained) )
```
